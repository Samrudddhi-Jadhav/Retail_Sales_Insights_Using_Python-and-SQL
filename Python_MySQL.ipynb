```python
#extract file from zhip file
import zipfile

with zipfile.ZipFile('orders.csv.zip', 'r') as zip_ref:
    zip_ref.extractall()  # extract files to the current directory
zip_ref.close() #close file
```

```python
#read data from file and handle null values
import pandas as pd
#df = pd.read_csv("orders.csv")
df = pd.read_csv("orders.csv", na_values = ['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
# rename columns names....make them lower case and relace with underscore
#df.rename(columns = {'order Id' : 'order_id', 'City':'city'})
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.head(5)
```

```python
#drive new columns discount, sale price and profit
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df
```

```python
#convert order datefrom object data type to datetime
#df.dtypes
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
```

```python
#drop cost price list price and discount percent columns
df.dtypes
df.drop(columns=["list_price","cost_price","discount_percent"],inplace=True)
```
